# 8958
OX퀴즈

```python
a = int(input())
for i in range(a):
    score = 0 
    ans = 0 #다음 줄로 넘어가기전에 초기화
    quiz = list(input() #for문 조건문으로 쓰기위해서 리스트로 받음
    for i in quiz:
        if i == "O":
            score +=1  #연속으로 O가 나오면 1씩 추가
            ans +=score # 그 score를 ans에 더하기
        else:
            score =0 # 다시 O가 나오기전에 score값 초기화
    print(ans)
```

# 4344
평균은 넘겠지

```python
N = int(input())
for i in range(N):
    a = list(map(int,input().split()))
    b = a[0] #일단 첫번째값 따로 저장해두고
    a.remove(b) #첫번째값을 리스트에서 삭제
    mean = sum(a)/len(a) #평균 계산
    count = 0 #줄마다 리셋
    for i in range(len(a)):
        if a[i]>mean:
            count +=1
    percen = count/len(a)
    percen = percen*100
    ans = "{:.3f}%".format(percen) #이건 문법
    print(ans)
```

# 15596
정수 N개의 합

```python
def solve(a):
    ans = 0
    for i in range(len(a)):
        ans += a[i]
    return ans
```

# 5622
다이얼

```python
a = input()
b = len(a)
ans = 0
for i in range(b):
    if a[i] in "ABC":
       ans+=3 
    elif a[i] in "DEF":
        ans+=4
    elif a[i] in "GHI":
        ans+=5
    elif a[i] in "JKL":
        ans+=6
    elif a[i] in "MNO":                
        ans+=7
    elif a[i] in "PQRS":
        ans+=8
    elif a[i] in "TUV":
        ans+=9
    elif a[i] in "WXYZ":
        ans+=10
print(ans)
```
